```python
#get quake data
import json
import requests
def call():
    url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson'
    data = requests.get(url)
    return data.json()
call()
```

```python
def get_quakes():
    quakes = data['features']
    return quakes
get_quakes()
```
